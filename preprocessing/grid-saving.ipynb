```python
!pip install netcdf4
import xarray as xr
import numpy as np
from huggingface_hub import hf_hub_download
```

```python
ds_grid_HF = hf_hub_download(repo_id = "LEAP/ClimSim_low-res",
                         filename="ClimSim_low-res_grid-info.nc",
                         repo_type="dataset") #Downloading dataset
ds_grid = xr.open_dataset(ds_grid_HF) #Opening dataset
```

```python
ds_grid
```

**I will save the variables in ds_grid as compressed binary arrays. It will avoid re-downloading from HuggingFace and to simplify loading in subsequent steps**

```python
area = np.asarray(ds_grid["area"])
hyai = np.asarray(ds_grid["hyai"])
hyam = np.asarray(ds_grid["hyam"])
hybi = np.asarray(ds_grid["hybi"])
hybm = np.asarray(ds_grid["hybm"])
lat  = np.asarray(ds_grid["lat"])
lon  = np.asarray(ds_grid["lon"])
lev  = np.asarray(ds_grid["lev"])
ilev = np.asarray(ds_grid["ilev"])
```

**Saving data as npz file**

```python
np.savez("grid_data.npz",area = area,
                        hyai = hyai,
                        hyam = hyam,
                        hybi = hybi,
                        hybm = hybm,
                        lat  = lat,
                        lon  = lon,
                        lev  = lev,
                        ilev = ilev)
```

## Loading created file

```python
grid_data_npz_files = np.load("grid_data.npz").files
```
