```python
def vertical_NO_vertical(cols_name_in,col_patterns):
    '''
    This function separates columns names into two categories, those with vertical dimensions and does without vertical
    dimension (scalar)
    Inputs:
    cols_name_in: list, total number column names
    col_patterns: list,pattern with the start name of the vertical variables
    Ouptus:
    X_col_series_out: list, columns names with vertical levels 
    X_col_series_NOT_out: list, columns names without vertical levels
    X_col_series_indices_out: array, columns indices with variables that have vertical levels 
    X_col_series_NOT_indices_out: array, columns indices with variables that DON'T have vertical levels 
    '''
    X_col_series_out = [x for x in cols_name_in if np.mean([x.startswith(y) for y in col_patterns])>0] #it will be true
                # only for when the column starts with one of the col_name_parent and the mean will be larger than 0

    X_col_series_NOT_out  = [x for  x in cols_name_in if x not in X_col_series_out ] #Select the columns that don't have vertical dimension
    
    #Checks if the number of columns with and without vertical levels is the same as the initial (combined) number of columns
    print(f"Sum of vertical and not vertical variables equal to initial number of columns: --> {len(X_col_series_out )+len(X_col_series_NOT_out ) == len(cols_name_in)}")
    
    #----------- Obtaining the inidices of the columns for each varaible clase ------------------
    #---------- variables with vertical dimensions
    X_col_series_indices_out = np.asarray([np.where(np.asarray(cols_name_in) == x)[0][0] for x in X_col_series_out ])
    #---------- variables withOUT vertical dimensions
    X_col_series_NOT_indices_out = np.asarray([np.where(np.asarray(cols_name_in) == x)[0][0] for x in X_col_series_NOT_out ])

    return(X_col_series_out, X_col_series_NOT_out, X_col_series_indices_out, X_col_series_NOT_indices_out)
```

```python
FEAT_COLS = train_df.columns[1:557] #Input features (columns names)
TARGET_COLS = train_df.columns[557:] #Targets (columns names)
##
#Selecting the variables that have vertical levels
col_name_parent = ['state_t', 'state_q0001', 'state_q0002', 
                   'state_q0003', 'state_u', 'state_v',
                   'pbuf_ozone', 'pbuf_CH4', 'pbuf_N2O'] # These are the basename of the variables with vertical levels
## Obtaining feature columns names and indices for variables with vertical and not vertical levels
(X_col_series,
    X_col_series_NOT,
    X_col_series_indices,
    X_col_series_NOT_indices,
) = vertical_NO_vertical(cols_name_in = FEAT_COLS, col_patterns = col_name_parent)

## ---------- Dividing target variables into variables with vertical leves and variables without vertical levels
y_name_parent = ['ptend_t', 'ptend_q0001', 'ptend_q0002', 'ptend_q0003', 'ptend_u', 'ptend_v']
(y_col_series,
    y_col_series_NOT,
    y_col_series_indices,
    y_col_series_NOT_indices,
) = vertical_NO_vertical(cols_name_in = TARGET_COLS, col_patterns = y_name_parent)
```

### Functions to select variables from Netcdf files

```python
#------------------- Function to get feature variables from nc file diving them according to ----------------
#------------------------ variables with vertical and not vertical dimension ----------------------
def get_data_features(nc,col_names,transpose):
    '''
    It goes through each colum in col_names and select the data from nc for that column.
    If transpose is true, it transpose the variables with vertical dimensions. The input dimension is
    (lev, ncol), after transposed it  is (ncol,lev). Append all variables and finally it concatenate them with the
    last dimension (columns). Hence, if there are 60 vertical level, and 10 variables with  vertical levels. The final
    dimension will be (ncol,60*10). In case it doesn't have vertical level, it creates a concatenate array of dimension
    (ncol, nvariables) where nvariables is  the number of variables with no vertical level.
    Inputs:
    nc: xr array with variables
    col_names: array, column names to iterate
    transpose: boolean, whether to transpose or not the select variable.
    '''
    data_cols = []
    for col in col_names:
        arr = np.asarray(nc[col]) # Selects the column in "col_names"
        if transpose:
            arr = np.transpose(arr, [1,0]) # Transpose from (lev, ncol) to (ncol,lev)
        data_cols.append(arr) # Append variables

    if transpose:
        data_out = np.concatenate(data_cols, axis=-1)
        #--checking correct dimensions
        shape_data_vertical_level = np.shape(data_out)
        if shape_data_vertical_level[0] != 384 or shape_data_vertical_level[1] != 540:
            raise Exception("Shape of data_vertical_level is not 384 or 540")
    elif transpose == False:
        data_out = np.stack(data_cols, axis=-1)
        #--checking correct dimensions
        shape_data_vertical_level = np.shape(data_out)
        if shape_data_vertical_level[0] != 384 or shape_data_vertical_level[1] != 16:
            raise Exception("Shape of data_vertical_level is not 384 or 16")

    return(data_out)

#------------------- Function to get targets from nc file diving them according to ----------------
#------------------------ variables with vertical and not vertical dimension ----------------------

def get_data_targets(nc_target,nc_features,col_names, transpose):
    '''
    It iterates through the elements in "col_names" to select that variable in "nc_target". Variables without
    vertical level are not transposed. They are concatenated and the output dimension is (ncol, nvariables) where
    nvariables is the number of variables with no vertical dimension. Ideally for the case of the target variables with 
    vertical dimension, the function would take each variable and concatenate them. Nevertheless, according to the
    discussion of this competition, these variables are computed as a delta, meaning that the real target value is
    y = (var_target-var_feature)/Delta_t, where var_target and var_feature correspond to the same variable name
    but in each nc file, and Delta_t = 1200 s. The output dimension is (ncol,60*nvariables_v)
    Inputs:
    nc_target: ncfile, file containing target variables
    nc_features: ncfile, file containing feature variables
    col_names: array, names to iterate through
    transpose: boolean, whether or not to trasnpose the selected variable
    '''
    data_cols = []
    for col in col_names:
        arr_target = np.asarray(nc_target[col]) #Selects the variable from the target dataframe
        if transpose:
            arr_feat = np.asarray(nc_features[col]) #Selects the same variable from the feature dataframe
            arr = (arr_target-arr_feat)/1200 #Performs the difference
            arr = np.transpose(arr,[1,0]) #tranpose from (lev, ncol) to (ncol,lev)
            
        elif transpose == False:
            arr = arr_target # Sets the output array as the variable from the target dataframe

        data_cols.append(arr) # Append the dataframe
    if transpose: #outside the loop of variables
        data_out = np.concatenate(data_cols, axis=-1) #concatenate along the columns (vertical levels)
        #--checking correct dimensions
        shape_data_vertical_level = np.shape(data_out)
        if shape_data_vertical_level[0] != 384 or shape_data_vertical_level[1] != 360:
            raise Exception("Shape of data_vertical_level is not 384 or 360")
    elif transpose == False:
        data_out = np.stack(data_cols, axis=-1) # Concatenate along the variables (no vertical dimension)
        #--checking correct dimensions
        shape_data_vertical_level = np.shape(data_out)
        if shape_data_vertical_level[0] != 384 or shape_data_vertical_level[1] != 8:
            raise Exception("Shape of data_vertical_level is not 384 or 8")
        
    return(data_out)     
```

### Iterating through several files


```python
from huggingface_hub import hf_hub_download, hf_hub_url, get_hf_file_metadata
import os

def safe_hf_download(repo_id, filename, repo_type="dataset", max_retries=3):
    """Download a file from HuggingFace Hub, verifying size and retrying if needed."""
    for attempt in range(max_retries):
        try:
            # Get expected file size from Hub metadata
            url = hf_hub_url(repo_id=repo_id, filename=filename, repo_type=repo_type)
            metadata = get_hf_file_metadata(url)
            expected_size = metadata.size  # in bytes

            # Download (uses cache if already present)
            local_path = hf_hub_download(
                repo_id=repo_id,
                repo_type=repo_type,
                filename=filename,
                force_download=(attempt > 0)  # force re-download on retries
            )

            # Verify size
            actual_size = os.path.getsize(local_path)
            if expected_size is not None and actual_size != expected_size:
                print(f"  ⚠️  Size mismatch for {filename}: "
                      f"expected {expected_size}, got {actual_size}. Retrying...")
                continue

            return local_path

        except Exception as e:
            print(f"  ❌ Attempt {attempt+1} failed for {filename}: {e}")
            if attempt == max_retries - 1:
                raise

    raise RuntimeError(f"Failed to download {filename} after {max_retries} attempts")
```

```python
#---------------------------- Defining initial variables -------------------------------------
space_dim = 384 # this corresponds to "ncol" in ncfiles. It assummes that every file has the same spatial dimension
vertical_dim = 60 # this corresponds to "lev" in ncfiles. It assummes that every file has the same vertical dimension
len_files = len(all_files_lowres) # number of files
N_vertical_features = len(col_name_parent) # Number of features with vertical levels
#---------------- Feature variables
feat_vertical = np.empty((space_dim*len_files,N_vertical_features*60))# Features with vertical levels. 
                                                # Dimensions: (NumberFiles*SpatialDimension,NumberVariablesVertical*VerticalDimension)
feat_NO_vertical = np.empty((space_dim*len_files, 16))  # Features with NO vertical levels.
                                                        # Dimensions (NumberFiles*SpatialDimension,NumberVariablesNoVertical)
#--------------- Target variables
N_vertical_target = len(col_name_parent[:6]) # Number of features with vertical levels
target_vertical = np.empty((space_dim*len_files,N_vertical_target*60)) # Target with vertical levels
                                                                       # Dimensions: (NumberFiles*SpatialDimension,NumberVariablesVertical*VerticalDimension)
target_NO_vertical = np.empty((space_dim*len_files,8))
#---------------------------- Loop trhough files ---------------------------------------------
for i,file_use in enumerate(tqdm(all_files_lowres)):
    #---------------------------- Downloading files ------------------------------------------
    #-- File containing feature variables
    rel_path = file_use.replace("datasets/LEAP/ClimSim_low-res/", "")
    target_rel_path = rel_path.replace('.mli.', '.mlo.')

    try:
        file_path = safe_hf_download("LEAP/ClimSim_low-res", rel_path)
        target_path = safe_hf_download("LEAP/ClimSim_low-res", target_rel_path)
    except Exception as e:
        print(f" Permanently failed: {rel_path} — {e}")
        failed_files.append((rel_path, target_rel_path))
        continue
    #-----------------------------------------------------------------------------------------
    #---------------------------- Reading downloaded files -----------------------------------
    df = xr.open_dataset(file_path, engine="netcdf4")
    df_target = xr.open_dataset(target_path, engine="netcdf4")
    #-----------------------------------------------------------------------------------------
    #---------------------------- Getting features and targets -------------------------------
    #--------- Features
    data_vertical_level = get_data_features(nc=df,col_names=col_name_parent,transpose=True)
    data_NO_vertical_level = get_data_features(nc=df,col_names=X_col_series_NOT,transpose=False)
    #--------- Targets
    data_target_vertical_level = get_data_targets(nc_target=df_target,nc_features=df,
                                              col_names=col_name_parent[:6], transpose=True)
    data_target_NO_vertical_level = get_data_targets(nc_target=df_target,nc_features=df,
                                              col_names=y_col_series_NOT, transpose=False)
    #-----------------------------------------------------------------------------------------
    #---------------------------- Adding obtained variables to containers --------------------
    #---------------- Feature variables
    feat_vertical[i*space_dim:(i+1)*space_dim,:] = data_vertical_level
    feat_NO_vertical[i*space_dim:(i+1)*space_dim,:] = data_NO_vertical_level
    #--------------- Target variables
    target_vertical[i*space_dim:(i+1)*space_dim,:] = data_target_vertical_level
    target_NO_vertical[i*space_dim:(i+1)*space_dim,:] = data_target_NO_vertical_level
```
